El método .search() solo regresa la primera aparicion del patron buscado en el texto como un objeto .Match

```python
resultado = re.search('la carencia',texto)

if resultado != None:
  print('Busqueda exitosa')
  print(resultado)
else:
  print('No se encontro el patron buscado')
```

El métod .findall ser regresa un objeto de tipo list con todas las apariciones del patron.

```python
resultado = re.findall('la carencia',texto)

if resultado != None:
  print('Busqueda exitosa')
  print(resultado)
else:
  print('No se encontro el patron buscado')
```

Simbolos especiales

Acento circunflejo (^)

Sirve para obtener la cadena que inicia con el patros escrito despues del simbolo "^"

```python
text = texto.split('\n')
for line in text:
  if re.search('^Pa',line):
    print(line)
```

Signo '$' en regex

Funciona para encontrar las cadenas que terminen en ese patron

```python
for line in text:
  if re.search('ar$',line):
    print(line)
```

Uso de corchetes []

Uso de [] en Expresiones Regulares

Los corchetes [] en regex definen un conjunto de caracteres permitidos. La expresión coincide con solo un carácter, pero ese carácter puede ser cualquiera de los incluidos dentro de los corchetes.

Funcionamiento básico

[abc] coincide con un solo carácter que sea a, b o c.

Ejemplos:

Coincide: a, b, c

No coincide: d, x, ab (dos caracteres)

Rangos de caracteres

Dentro de los corchetes se pueden definir rangos:

[a-z] → letras minúsculas

[A-Z] → letras mayúsculas

[0-9] → dígitos

Ejemplo combinado:

[a-zA-Z0-9]


Coincide con cualquier letra o número.

Negación

Si el primer carácter dentro de los corchetes es ^, el conjunto se niega:

[^abc] → cualquier carácter que no sea a, b o c

Caracteres especiales dentro de []

Algunos caracteres pierden su significado especial dentro de corchetes, pero hay excepciones:

] → si lo quieres incluir, colócalo al inicio: []abc]

- → define rangos; para usarlo literal, colócalo al inicio o al final: [-abc] o [abc-]

^ → solo significa negación si está en la primera posición

Ejemplos útiles

Coincidir vocales:

[aeiou]


Coincidir letra o dígito:

[a-zA-Z0-9]


Coincidir cualquier carácter que no sea un dígito:

[^0-9]

Coincidir con cualquier caracter:

[.]
