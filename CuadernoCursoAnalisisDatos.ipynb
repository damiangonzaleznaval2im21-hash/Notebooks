```python
import random

nombres_base = ["Ana", "Pedro", "Juan", "Joseline", "Luis", "Maria", "Carlos", "Sofia"]
ciudades_base = ["Paris", "CDMX", "Roma", "Barcelona", "Berlin", "Tokyo"]
edades_base = [18, 20, 22, 25, 30, 35, 40, 42]

datos = {
    "Nombres": [random.choice(nombres_base) for _ in range(100)],
    "Edad": [random.choice(edades_base) for _ in range(100)],
    "Ciudad": [random.choice(ciudades_base) for _ in range(100)]
}
```

```python
df = pd.DataFrame(data = datos)
df[(df['Edad'] < 30) & (df['Ciudad'] == "CDMX")].Nombres.count()
```
